# STL -> Seasonal and Trend decomposition using Loess

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import warnings
warnings.filterwarnings('ignore')

from stldecompose import decompose, forecast
```

```python
df = pd.read_csv('../datasets/AAPL.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='ignore') 
df.head()
```

# Using STLDecompose

```python
df_2018 = df[df['Date'] > datetime.datetime(2014,12,31,23,59)]
```

```python
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], df_2018['Close'])
plt.xticks(rotation=90)
plt.show()
```

```python
temp = df_2018.reset_index()
df_stl_month = temp[['index', 'Close']].set_index('index')
decomp = decompose(df_stl_month.values, period=48)
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

## Approximate trend

```python
print("Find the Approximate Trend line which fits the above time series data.")
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], trend, label='trend')
plt.plot(df_2018['Date'], df_2018['Close'])
plt.title("Approximate Trend")
plt.xlabel("Date")
plt.ylabel("Trend")
plt.legend()
plt.show()
```

## Seasonality

```python
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], seasonal, label='seasonal')
plt.title("Seasonal")
plt.xlabel("Date")
plt.ylabel("Seasonal")
plt.legend()
plt.show()
```

## Residual

```python
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], residual, label='residual')
plt.title("Residual")
plt.xlabel("Date")
plt.ylabel("Residual")
plt.legend()
plt.show()
```

## Detrended Series vs Residual

```python
detrended_series = df_2018['Close'] - trend

plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], detrended_series, marker='', linestyle='-', label='detrended_timeseries')
plt.plot(df_2018['Date'], residual, label='residual')
plt.title("Detrended Series vs Residual")
plt.xlabel("Date")
plt.ylabel("value")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
print("Find the De-seasonalised time series by subtracting time series data with seasonal data")

deseasonalised = df_2018['Close'] - seasonal
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], deseasonalised, marker='', linestyle='-', label='deseasonalised')
plt.plot(df_2018['Date'], df_2018['Close'])
plt.title("Deseasonalited vs Trend")
plt.xlabel("Date")
plt.ylabel("value")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# Using Linear Regressor

```python
# importing libraries
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

# index
X = temp.index.values
# timeseries value
y = temp.Close.values

# scaling
scaler = MinMaxScaler()
scaler = scaler.fit(y.reshape(-1,1))
y_scaled = scaler.transform(y.reshape(-1,1))

# creating poly feature of degree 5
poly = PolynomialFeatures(5)
X_poly = poly.fit_transform(X.reshape(-1,1))

# fitting model
model = LinearRegression()
model = model.fit(X_poly, y_scaled)

# approx_trend 
approx_trend = model.predict(X_poly)
temp['approx_trend'] = scaler.inverse_transform(approx_trend)

# plotting
plt.figure(figsize=(30,10))
plt.rcParams.update({'font.size': 22})
plt.grid()
plt.plot(df_2018['Date'], y, marker='', linestyle='-', label='Close Value', linewidth=4)
plt.plot(df_2018['Date'], temp.approx_trend, marker='', linestyle='-', label='approx_trend', linewidth=4)
plt.plot(df_2018['Date'], trend, label='trend (STL)', linewidth=4)
plt.title("")
plt.xlabel("Date")
plt.ylabel("Close")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
